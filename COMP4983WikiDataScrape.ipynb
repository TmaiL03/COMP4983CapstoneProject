```python
# Initializing the main dataframe.
database = pd.DataFrame(columns = ["year", "round", "gpName", "circLen", "propLaps", "raceLen", "weather", "driver", "team", "grid", "finish", "dnf"])
```

```python
# Defining a function to find data in race overview table.
def fetchOverviewData(df, key):

  # Iterate over each row of the provided table until key is found in the first column.
  for index, row in df.iterrows():

    # If first column of row is key, return adjacent value.
    if(row.iloc[0] == key):

      return row.iloc[1]
      break

    else:
      pass

  # If no data is located for the provided key, inform the user.
  print(f"No data found for {key} for {df.iloc[1,1]}")
```

```python
# Defining function to iteratively find the race classification table.
def findWikiRaceTable(url):

  # Reading in data from URL.
  tables = pd.read_html(url)
  raceLaps = tables[0].iloc[9,1].split(" ")[0]

  # Iterate over each table to check if the leader's number of laps equals the listed race distance.
  for table in tables:

    # Check if first column name in table is data type string.
    if(table.columns[0].__class__.__name__ == "str"):

      # Check if column name matches "Laps" or "Laps1" ("Laps1" is necessary in order to handle footnotes).
      if(("Laps" in table.columns) or ("Laps1" in table.columns)):

        # Check if first 2 characters of string equal value of listed race laps.
        if(str(table.iloc[0,4])[:2] == raceLaps):
          return table

        else:
          pass

      else:
        pass

    else:
      pass
```

```python
# Defining function to append season data to database.
def appendSeason(database, seasonURLs):

  # Iterate over each race from the provided season's URL list.
  for url in seasonURLs:

    # Retrieving and organizing tables.
    tables = pd.read_html(url)
    general = tables[0]

    raceData = findWikiRaceTable(url)

    # General race information to be reused for each driver entry.
    raceInfo = {}
    raceInfo["Year"] = general.columns[0].split(" ")[0]
    raceInfo["Round"] = seasonURLs.index(url) + 1
    raceInfo["Grand Prix Name"] = url.split(f"{raceInfo['Year']}_")[1].replace("_", " ")
    raceInfo["Circuit Length"] = float(general.iloc[8,1].split(" ")[0])
    raceInfo["Number of Laps"] = int(fetchOverviewData(general, "Distance")[:2])
    raceInfo["Race Length"] = float(raceInfo["Circuit Length"]) * raceInfo["Number of Laps"]
    weather = fetchOverviewData(general, "Weather")

    # Reducing dimensionality of weather attribute.
    if(("rain" in weather.lower()) or ("wet" in weather.lower())):
      raceInfo["Weather"] = "Rain"
    else:
      raceInfo["Weather"] = "Dry"

    # Omitting records from each race result table included the following in the "Pos." field.
    omit = ["Fastest", "Source", "107%", "WD", "DNS"]

    # Iterate over each row of the raceData table.
    for driver in range(len(raceData)):

      # If row contains any of the terms in "omit", ignore this row.
      if any(s in raceData.iloc[driver, 0] for s in omit):
        pass

      else:
        driverInfo = {}
        driverInfo["Driver Name"] = raceData.iloc[driver, 2]
        driverInfo["Team Name"] = raceData.iloc[driver, 3]
        driverInfo["Laps Completed"] = float(int(raceData.iloc[driver, 4]) / raceInfo["Number of Laps"])
        driverInfo["Finishing Position"] = driver + 1

        # If a driver starts from the pitlane ("PL"), then assign "Grid Position" of 20, else.
        if(raceData.iloc[driver, 6].startswith("PL")):
          driverInfo["Grid Position"] = "20"

        # Handles single-digit grid positions (other than 1st) with footnotes attached to them.
        elif((int(raceData.iloc[driver, 6]) > 20) and (int(raceData.iloc[driver, 6]) <= 100)):
          driverInfo["Grid Position"] = str(raceData.iloc[driver, 6])[:1]

        # Handles double-digit grid positions with footnotes attached to them.
        elif(int(raceData.iloc[driver, 6]) > 100):
          driverInfo["Grid Position"] = str(raceData.iloc[driver, 6])[:2]

        else:
          driverInfo["Grid Position"] = raceData.iloc[driver, 6]

        # If a driver's "Time/Retired" column's first character is not a number or a "+" sign, DNF = True, else False.
        if(raceData.iloc[driver, 5][0].isdigit() or raceData.iloc[driver, 5][0] == "+"):
          driverInfo["DNF"] = False
        else:
          driverInfo["DNF"] = True

        # Appending the compiled entry to the database.
        database.loc[len(database)] = [
            raceInfo["Year"],
            raceInfo["Round"],
            raceInfo["Grand Prix Name"],
            raceInfo["Circuit Length"],
            driverInfo["Laps Completed"],
            raceInfo["Race Length"],
            raceInfo["Weather"],
            driverInfo["Driver Name"],
            driverInfo["Team Name"],
            driverInfo["Grid Position"],
            driverInfo["Finishing Position"],
            driverInfo["DNF"]
        ]
```

```python
# Defining function to handle simplifying team names in an existing Pandas DataFrame.
def simplifyTeamNames(df):

  # Simplified names.
  simpleNames = ["Alpine", "Aston Martin", "Ferrari", "Haas", "McLaren", "Mercedes", "RB", "Red Bull", "Sauber", "Williams"]

  # Ferrari and Mercedes needn't be handled as their listed names are already simplified.

  # Rule 1: Handling Alpine, Aston Martin, Haas, McLaren, Red Bull, and Williams (and partially RB) will require checking if the team name starts with the simplified variant.
  rule1Names = ["Alpine", "Aston Martin", "Haas", "McLaren", "RB", "Red Bull", "Williams"]

  # Rule 2: Handling RB (specifically the AlphaTauri variant) and Sauber will require specific instances (their various names possess no immediate resemblance to their simplified versions).
  rule2NamesSauber = ["Alfa", "Kick"]
  rule2NamesRB = ["AlphaTauri"]


  # Filter through each team name in provided DataFrame df.
  for index, row in df.iterrows():

    # Handling Rule 1.
    for r1name in rule1Names:
      if(row["team"].startswith(r1name)):
        row["team"] = r1name
        df.at[index, "team"] = row["team"]

      else:
        pass

    # Handling Rule 2a: Converting "AlphaTauri"-related instances to "RB".
    if(row["team"].startswith(rule2NamesRB[0])):
      row["team"] = "RB"
      df.at[index, "team"] = row["team"]

    # Handling Rule 2b: Converting applicable instances defined in rule2NamesSauber to "Sauber".
    if(row["team"].startswith(rule2NamesSauber[0]) or row["team"].startswith(rule2NamesSauber[1])):
      row["team"] = "Sauber"
      df.at[index, "team"] = row["team"]

    else:
      pass

  return df
```

```python
# Iterate over each race from the 2022 season.
appendSeason(database, season2022)

# Iterate over each race from the 2023 season.
appendSeason(database, season2023)

# Iterate over each race from the 2024 season.
appendSeason(database, season2024)

# Simplify team names.
simplifyTeamNames(database)

# Printing to examine database.
database
```
